```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
test_acc = [95.07, 95.97, 96.27, 96.28, 96.62, 96.23, 96.95]
robust_acc = [86.5, 87.57, 81.52, 89.62, 86.4, 84.05, 88.69]
method = ['Vanilla','Mixup','CutMix','Manifold','Puzzle Mix','Co-Mixup','SMART']

fig, ax = plt.subplots(figsize=(10,5))

ax.add_patch(Rectangle((95.07,86.5), 10, -10, alpha = 0.1, color='C3',label='Robustness Trade-off'))
ax.add_patch(Rectangle((95.07,86.5), 10, 10, alpha = 0.1, color='C2',label='Improve OOD Generalization'))
ax.scatter(test_acc[0:-1], robust_acc[0:-1], s=60, color='C0')
ax.scatter(test_acc[-1], robust_acc[-1], s=120, color='C1',marker='*')

ax.set_ylim([80.5, 91])
ax.set_xlim([95, 97.5])
ax.tick_params(axis='both', which='major', labelsize=14)

for i, txt in enumerate(method):
    ax.annotate(txt, (test_acc[i]+0.05, robust_acc[i]+0.05),fontsize = 20)

ax.set_ylabel('Average robust accuracy (%)', fontsize = 15)
ax.set_xlabel('Standard test accuracy (%)', fontsize = 15)
# ax.set_title('Trade-off between robustness and standard accuracy (CIFAR-10)', fontsize = 17)
# plt.legend(fontsize = 12, loc=4)
ax.grid()

```

```python
test_acc = [76.8, 77.4, 78.96, 78.51, 79.65, 79.68, 79.91]
robust_acc = [55.71, 56.93, 49.88, 55.01, 51.37, 51.44, 56.21]
method = ['Vanilla','Mixup','CutMix','Manifold','Puzzle Mix','Co-Mixup','SMART']

fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(test_acc[0:-1], robust_acc[0:-1], s=60, color='C0')
ax.scatter(test_acc[-1], robust_acc[-1], s=120, color='C1',marker='*')
ax.add_patch(Rectangle((76.8,55.71), 10, -10, alpha = 0.1, color='C3',label='Robustness Trade-off'))
ax.add_patch(Rectangle((76.8,55.71), 10, 10, alpha = 0.1, color='C2',label='Improve OOD Generalization'))

ax.set_ylim([49, 58])
ax.set_xlim([76.68, 80.75])
ax.tick_params(axis='both', which='major', labelsize=14)

for i, txt in enumerate(method):
    if txt == 'Puzzle Mix':
        ax.annotate(txt, (test_acc[i]-0.8, robust_acc[i]+0.1),fontsize = 20)
    elif txt == 'Co-Mixup':
        ax.annotate(txt, (test_acc[i]+0.1, robust_acc[i]+0.1),fontsize = 20)
    else:
        ax.annotate(txt, (test_acc[i]+0.1, robust_acc[i]+0.1),fontsize = 20)

ax.set_ylabel('Average robust accuracy (%)', fontsize = 15)
ax.set_xlabel('Standard test accuracy (%)', fontsize = 15)
# ax.set_title('Trade-off between robustness and standard accuracy (CIFAR-100)', fontsize = 17)
plt.legend(fontsize = 12, loc=3)
ax.grid()
```
