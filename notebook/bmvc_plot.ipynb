```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
test_acc = [95.07, 95.97, 96.27, 96.28, 96.62, 96.23, 96.95]
robust_acc = [86.5, 87.57, 81.52, 89.62, 86.4, 84.05, 88.69]
method = ['Vanilla','Mixup','CutMix','Manifold','Puzzle Mix','Co-Mixup','SAGE']

fig, ax = plt.subplots(figsize=(10,3))

ax.add_patch(Rectangle((95.07,86.5), 10, 10, alpha = 0.1, color='C2',label='Improve OOD Generalization'))
ax.add_patch(Rectangle((95.07,86.5), 10, -10, alpha = 0.1, color='C3',label='Robustness Trade-off'))

ax.scatter(test_acc[0:-1], robust_acc[0:-1], s=60, color='C0')
ax.scatter(test_acc[-1], robust_acc[-1], s=120, color='C1',marker='*')

ax.set_ylim([80.5, 91])
ax.set_xlim([95, 97.25])
ax.tick_params(axis='both', which='major', labelsize=14)

for i, txt in enumerate(method):
    ax.annotate(txt, (test_acc[i]+0.03, robust_acc[i]+0.03),fontsize = 17)

ax.set_ylabel('Average robust accuracy (%)', fontsize = 11)
ax.set_xlabel('Standard test accuracy (%)', fontsize = 14)
# ax.set_title('Trade-off between robustness and standard accuracy (CIFAR-100)', fontsize = 17)
plt.legend(fontsize = 14, loc=3)
ax.grid()
plt.tight_layout()
plt.savefig('/h/ama/workspace/ama-at-vector/best-mix/robustness-cifar10.pdf') 
# plt.savefig('/h/ama/workspace/ama-at-vector/best-mix/robustness-cifar10.png', format="png", dpi=500)
```
